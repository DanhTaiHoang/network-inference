# Maximum Likelihood Estimation 

[![Binder](https://mybinder.org/badge.svg)](https://mybinder.org/v2/gh/danhtaihoang/network-inference/blob/master/sphinx/codesource/mle.ipynb/master)

Maximum likelihood estimation (MLE) is a standard method in the field of network inference.

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import simulate
import inference
%matplotlib inline

np.random.seed(1)
```

Similary, we first consider a system of $N = 100$ variables, coupling variance parameter $g = 2.0$

```python
# parameter setting:
n = 100    # number of variables
g = 2.0    # coupling variance
w0 = np.random.normal(0.0,g/np.sqrt(n),size=(n,n))
```

Data length $L=2000$ is used as the first test.

```python
l = 2000
s = simulate.generate_data(w0,l)
```

Similar to FEM and eMF, MLE is an iterative algorithm. We first run this method as discribed in (Ref), without stopping criterion.

```python
niter = 1000
rate = 1.
w = inference.mle(s,niter,rate,stop_criterion='no')
```

```python
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.title('actual coupling matrix')
plt.imshow(w0,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,1))
plt.title('predicted coupling matrix')
plt.imshow(w,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,2))
plt.title('predicted couplings vs. actual couplings')
plt.plot([-1,1],[-1,1],'r--')
plt.scatter(w0,w)
plt.xlabel('actual couplings')
plt.ylabel('inferred couplings')

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.show()
```

```python
MSE = np.mean((w-w0)**2)
print('MSE:',MSE)
```

Now, we apply our stopping criterion for this method. We stop the iteration when the discrepancy starts to increase.  We will see that the performance significantly improve.

```python
w = inference.mle(s,niter,rate,stop_criterion='yes')
```

```python
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.title('actual coupling matrix')
plt.imshow(w0,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,1))
plt.title('predicted coupling matrix')
plt.imshow(w,cmap='rainbow',origin='lower')
plt.xlabel('j')
plt.ylabel('i')
plt.clim(-0.5,0.5)
plt.colorbar(fraction=0.045, pad=0.05,ticks=[-0.5,0,0.5])

plt.subplot2grid((1,3),(0,2))
plt.title('predicted couplings vs. actual couplings')
plt.plot([-1,1],[-1,1],'r--')
plt.scatter(w0,w)
plt.xlabel('actual couplings')
plt.ylabel('inferred couplings')

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.show()
```

```python
MSE = np.mean((w-w0)**2)
print('MSE:',MSE)
```
