# Inferring coupling strengths from configurations of variables

```python
import numpy as np
import sys
import matplotlib.pyplot as plt

import fit_binary as fit
import simulate
import plot as pl

np.random.seed(1)

# parameter:
n = 100    # number of variables
```

# 1. Large sample size, $L = N^2$

```python
l = n**2
```

## 1.1. Small coupling variance, g = 0.5 

```python
# simulate data
g = 0.5
w0,s0 = simulate.simulate_data(n,g,l)
```

### 1.1.1. Naive Mean Field approach (nMF)

```python
w = fit.nmf(s0)
pl.plot_w(w0,w)
```

### 1.1.2. Thouless-Anderson-Palmer approach (TAP)

```python
w = fit.tap(s0)
pl.plot_w(w0,w)
```

### 1.1.3. Exact Mean Field approach (eMF)

```python
w = fit.tap(s0)
pl.plot_w(w0,w)
```
